```python
import numpy as np
import matplotlib.pyplot as plt
import h5py as h5
```

```python
# Data contains 2D[meshgrid] for each Mz Z, KDE obs, intrinsic on M and Z and Pdet on same values
#Pdet is computed by interpolation as we have Pdet at some points 
#and we use interpolation specially we get zeros for points not in posteriors for popIII-model 
data = h5.File("correct_Mz_Zz_KDE_int_ob_Pdet.hdf5", "r") 
Mtot = data['Mtot'][...]
Z = data['Z'][...]
KDEint = data['kdeint'][...]
KDEobs = data['kdeobs'][...]
Pdet = data['Pdet'][...]
data.close()
Savepdet = Pdet.copy()
```

```python
def twoDplot(Mm, Zz, Kde1, Kde2, model='popIII', title='observed'):
    fig, (ax1,ax2) = plt.subplots(ncols=2, nrows=1, figsize=(14, 5))
# Make the contours
    ctr1 = ax1.contourf(Mm, Zz, Kde1, cmap="viridis",)
# Create a colorbar
    cbar1 = fig.colorbar(ctr1, ax=ax1)
# Label the axes and colorbar
    ax1.set_xlabel("$Log10[Mz] \\, [M_\\odot]$", fontsize=16)
    ax1.set_ylabel("$z$", fontsize=16)
#cbar.set_label("$Kde(log(Mz), z)$")
    cbar1.set_label("$Pdf_{int}(Log10[Mz], z)$", fontsize=12)
    #ax1.set_title('KDE1')
    #plt.savefig("TwoDKDE"+title+model+".png")
    ctr2 = ax2.contourf(Mm, Zz, Kde2, cmap="viridis",)
# Create a colorbar
    cbar2 = fig.colorbar(ctr2, ax=ax2)
# Label the axes and colorbar
    ax2.set_xlabel("$Log10[Mz] \\, [M_\\odot]$", fontsize=16)
    ax2.set_ylabel("$z$", fontsize=16)
#cbar.set_label("$Kde(log(Mz), z)$")
    cbar2.set_label("$Pdf_{obs}(Log10[Mz], z)$", fontsize=12)
# Show the plot
    plt.show()
twoDplot(np.log10(Mtot), Z, KDEobs, KDEint, title='observed')
```

```python
import matplotlib.cm, matplotlib.colors
from matplotlib.colors import LogNorm
def plot2Dvirdis(XXval, YYval, ZZ1val, ZZ2val):
    fig, (axl, axr) = plt.subplots(nrows=1, ncols=2,figsize=(12, 5))
    p = axl.pcolormesh(XXval, YYval, ZZ1val, cmap=plt.cm.get_cmap('cividis'), norm=LogNorm())
    CS = axl.contour(XXval, YYval, ZZ1val, colors='white', linestyles='dashed', linewidths=2,norm=LogNorm())
    cbar = plt.colorbar(p, ax=axl)
    cbar.ax.tick_params(labelsize=20)
    axl.tick_params(labelsize=15)
    axl.set_xlabel(r'$Log10[Mz]$')
    axl.set_title("intrinsic")
    scale_y = 1
    #ticks_y = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/scale_y))
    #axl.yaxis.set_major_formatter(ticks_y)
    axl.set_ylabel(r"$z$", fontsize=20)
    pr = axr.pcolormesh(XXval, YYval, ZZ2val, cmap=plt.cm.get_cmap('cividis'), norm=LogNorm())
    CSr = axr.contour(XXval, YYval, ZZ2val, colors='white', linestyles='dashed', linewidths=2,norm=LogNorm())
    cbarR = plt.colorbar(pr, ax=axr)
    cbarR.ax.tick_params(labelsize=20)
    axr.tick_params(labelsize=15)
    axr.set_xlabel(r'$Log10[Mz]$')
    axr.set_title('detected')
    scale_y = 1
    #ticks_y = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/scale_y))
    #axr.yaxis.set_major_formatter(ticks_y)
    axr.set_ylabel(r"$z$", fontsize=20)

    
    fig.tight_layout()
    fig.suptitle("popIII model")
    return 0

plot2Dvirdis(np.log10(Mtot), Z, Rateint, Rateobs)
```

```python
def calculate_sma(data, window_size):
    sma = np.convolve(data, np.ones(window_size)/window_size, mode='valid')
    return sma
```
